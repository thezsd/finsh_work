## 只处理比fadb高的，除此以外什么聚类什么是否存在于ncbi中都不重要，都不管

```python
# 读取csv文件
import pandas as pd

df = pd.read_csv('/mnt/e/SDT/output/CORASON/modify_fabB1_fadb_nr_rep_0.7_0.7_rep_seq/result.csv')
df
```

# 抽取比fadb结合能高的，只保留比fadb低的，但是对于比较更加的细致一些

```python
# 获取第一行第3列到最后一列的值作为阈值
thresholds = df.iloc[0, 2:]

# 筛选第3列到最后一列中每一列的值都不大于其对应阈值的行
df = df[df.iloc[:, 2:].apply(lambda x: x <= thresholds, axis=1).all(axis=1)]
df
```

# 加载名字

```python
# 读入匹配表
df_pair = pd.read_csv('/mnt/e/SDT/output/CORASON/modify_fabB1_fadb_nr_rep_0.7_0.7_rep_seq/prefix_mapping.csv')
df_pair
```

```python
# 遍历df的行
df['ncbi'] = ''
for index, row in df.iterrows():
    # 遍历df_pair的行
    for index2, row2 in df_pair.iterrows():
        # 如果df的protein和df_pair的prefix相同
        if row['protein'] == row2['New Prefix']:
            # 将df的protein替换为df_pair的NCBI
            df.at[index, 'ncbi'] = row2['Old Prefix']

df
```

# 合并kactkm以及kcat结果

```python
kcatkm = pd.read_csv('/mnt/e/UniKP/Kcat_Km/pre_corason_all.csv')
kcatkm
```

```python
not_matched_ncbi = []

i = 0
for ncbi in df['ncbi']:
    match_found = False
    for pro in kcatkm['pro']:
        if ncbi in pro:  # 注意这里修改为了精确匹配 "==", 前面的"in"用法在原问题中可能不准确
            i += 1
            match_found = True
            break
    if not match_found:
        not_matched_ncbi.append(ncbi)

print("未找到匹配的ncbi物种名：", not_matched_ncbi)
```

## 筛选比fadb的kcatkm高的物种

```python
import pandas as pd

# 假设df和kcatkm已经被正确读取
# df = pd.read_csv('your_df.csv')
# kcatkm = pd.read_csv('kcatkm.csv')

# 确保处理潜在的NaN值
df.dropna(subset=['ncbi'], inplace=True)
kcatkm.dropna(subset=['pro'], inplace=True)

# 初始化一个新的DataFrame来存储满足条件的行
filtered_df = pd.DataFrame()

# 对于df中的每一行，检查ncbi列的值是否在kcatkm的pro列中，并且对应的kcatkm指标是否满足条件
for index, row in df.iterrows():
    # 直接检查ncbi的值是否为pro列中的子字符串，避免正则表达式错误
    matching_rows = kcatkm[kcatkm['pro'].apply(lambda x: row['ncbi'] in x)]
    
    # 如果找到匹配且对应的指标大于阈值，则保留该行
    if not matching_rows.empty:
        # 提取kcatkm的指标列
        kcatkm_values = matching_rows.iloc[:, 7:13]  # 从第8列到第13列
        # 检查所有指标是否都大于指定阈值
        if (kcatkm_values > [69.506558, 65.153952, 67.012397, 60.158627, 60.086514, 52.923152]).all(axis=1).any():
            # 将当前行添加到filtered_df中
            filtered_df = pd.concat([filtered_df, row.to_frame().T])
    if matching_rows.empty:
        print(f"No match found for ncbi: {row['ncbi']}")
# 最后，filtered_df将只包含那些匹配且kcatkm指标大于指定阈值的行
filtered_df
```

## 把kcatkm数值加入到列表中

```python
import pandas as pd

# 假设 df 和 kcatkm 是你的两个 DataFrame，包含 'ncbi' 和 'pro' 列

# 初始化一个新的 DataFrame 用于存储匹配的行
matched_rows = pd.DataFrame(columns=kcatkm.columns)
df = filtered_df
# 遍历 df 中的每一行
for index, row in df.iterrows():
    ncbi_value = row['ncbi']
    
    # 检查是否有匹配的 ncbi 值在 kcatkm 的 'pro' 列中
    matched_rows = kcatkm[kcatkm['pro'].str.contains(ncbi_value,regex=False)]
    
    # 如果有匹配，将匹配的行添加到 df 中对应的行后面
    if not matched_rows.empty:
        for _, match_row in matched_rows.iterrows():
            df.loc[index, matched_rows.columns] = match_row
    else :
        print(f"{ncbi_value}")
# 如果需要，重新索引 df
# df.reset_index(drop=True, inplace=True)
df 
```

```python
import pandas as pd
from Bio import SeqIO

# 假设df已经读取并定义
# df = pd.read_csv('/path/to/your/csv/file.csv')

# 读取fasta文件
fasta_file_path = "/mnt/e/SDT/output/CORASON/modify_fabB1_fadb_nr_rep_0.7_0.7_rep_seq/modify_fabB1_fadb_nr_rep_0.7_0.7_rep_seq.fasta"
output_fasta_path = "/mnt/e/SDT/output/CORASON/modify_fabB1_fadb_nr_rep_0.7_0.7_rep_seq/result_all.fasta"

# 读取fasta文件中的所有序列
sequences = SeqIO.parse(fasta_file_path, "fasta")

# 创建一个列表来存储匹配的序列
matched_sequences = []

# 遍历df的ncbi列
for ncbi in df['ncbi']:
    for record in sequences:
        if ncbi in record.id:
            matched_sequences.append(record)
            break  # 找到匹配的序列后跳出循环

# 将匹配的序列写入新的fasta文件
SeqIO.write(matched_sequences, output_fasta_path, "fasta")

print(f"匹配的序列已保存到 {output_fasta_path}")
```
